```python
import os
import base64
import requests
from PIL import Image
from io import BytesIO

# API endpoint URLs
llava_url = 'http://8.12.5.48:11434/api/generate'
runwayml_url = 'https://api-inference.huggingface.co/models/Adrenex/fastgen'

# Your Hugging Face token for authentication
hugging_face_token = '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'

# Initialize variables
llava_suggestion = ''
change_dress = False
first_image_link = None  # Variable to store the link of the first image
modifications = []  # List to store user-specified modifications

# Prompt the user for an occasion
occasion = input("Enter the occasion for dressing up: ")

while True:
    # Prompt the user for a question or exit option
    user_input = input("Ask the model a question about the dress or type 'exit' to end the chat: ")

    # Check if the user wants to exit
    if user_input.lower() == 'exit':
        print("Exiting the chat.")
        break

    # Prompt the user to select the type of suggestions they want
    print("Select the type of suggestions you want:")
    print("1. Outfit")
    print("2. Accessories")
    print("3. Shoes")
    print("4. Handbags")
    option = input("Enter the option number: ")

    # Add selected options to the LLAVA prompt
    selected_options = []
    if '1' in option:
        selected_options.append('outfit')
    if '2' in option:
        selected_options.append('accessories')
    if '3' in option:
        selected_options.append('shoes')
    if '4' in option:
        selected_options.append('handbags')

    # Add a string before the user's question for LLAVA prompt
    llava_prompt = f'''For the occasion "{occasion}", suggest outfits, accessories, shoes, and handbags.
    Give suggestions for: {", ".join(selected_options)} based on the following input: {user_input}'''

    if first_image_link is None:
        # Read and encode the first image in base64
        first_image_path = os.getcwd() + '\\dress_B0006HJ91S.jpg'
        with open(first_image_path, 'rb') as first_image_file:
            first_image_data = base64.b64encode(first_image_file.read()).decode('utf-8')
        first_image_link = first_image_data  # Store the link for the first image

    # Apply previous modifications to the LLAVA prompt
    modified_prompt = llava_prompt
    for mod in modifications:
        modified_prompt += f"\n{mod}"

    # Payload data for LLAVA model with image input
    llava_payload = {
        "model": "llava:7b-v1.6-mistral-q5_K_M",
        "prompt": modified_prompt,
        "stream": False,
        "images": [first_image_link]  # Use the link of the first image as input
    }

    # Headers with Hugging Face token for authentication
    headers = {
        'Authorization': f'Bearer {hugging_face_token}'
    }

    # Send POST request to LLAVA model with authentication headers
    llava_response = requests.post(llava_url, json=llava_payload, headers=headers)

    # Check response status from LLAVA model
    if llava_response.status_code == 200:
        # Get the LLAVA model's response
        llava_response_data = llava_response.json()
        llava_suggestion = llava_response_data['response']
        print(llava_response_data['response'])
        # Check if the user wants to change the dress based on LLAVA suggestion
        change_dress_input = input("Do you want to change your dress according to LLAVA's suggestion? (yes/no): ")

        if change_dress_input.lower() == 'yes':
            change_dress = True

    else:
        print(f"Error accessing the LLAVA model: {llava_response.status_code} - {llava_response.text}")

    # Check if the user wants to change the dress and there is a suggestion from LLAVA
    if change_dress and llava_suggestion:
        # Prompt the user for modifications to add
        modification_input = input("Specify the modifications you want to add: ")

        # Add the modifications input to the list of modifications
        modifications.append(modification_input)

    # Check if there is a suggestion from LLAVA after adding modifications
    if llava_suggestion:
        # Payload data for Diffusion model
        diffusion_payload = {
            "inputs": llava_suggestion
        }
        
        # Send POST request to Diffusion model
        diffusion_response = requests.post(runwayml_url, json=diffusion_payload, headers=headers)

        if diffusion_response.status_code == 200:
            print("Image generated successfully!")

            # Display the generated image
            generated_image = Image.open(BytesIO(diffusion_response.content))
            generated_image.show()

        else:
            print(f"Error generating image with Diffusion model: {diffusion_response.status_code} - {diffusion_response.text}")
```

```python
import os
import base64
import requests
from PIL import Image
from io import BytesIO
import time  # Import time module

# API endpoint URLs
llava_url = 'http://8.12.5.48:11434/api/generate'
runwayml_url = 'https://api-inference.huggingface.co/models/Adrenex/fastgen'

# Your Hugging Face token for authentication
hugging_face_token = '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'

# Initialize variables
llava_suggestion = ''
change_dress = False
first_image_link = None  # Variable to store the link of the first image
modifications = []  # List to store user-specified modifications

def check_model_availability():
    availability_url = 'https://api-inference.huggingface.co/models/Adrenex/fastgen'
    try:
        response = requests.head(availability_url)
        if response.status_code == 200:
            return True
    except Exception as e:
        print(f"Error checking model availability: {e}")
    return False

while True:
    # Prompt the user for a question or exit option
    user_input = input("Press any key to get suggestions about your outfit or type 'exit' to end the chat: ")

    # Check if the user wants to exit
    if user_input.lower() == 'exit':
        print("Exiting the chat.")
        break

    # Prompt the user to select the type of suggestions they want
    print("Select the type of suggestions you want:")
    print("1. Outfit")
    print("2. Accessories")
    print("3. Shoes")
    print("4. Handbags")
    option = input("Enter the option number: ")

    # Add selected options to the LLAVA prompt
    selected_options = []
    if '1' in option:
        selected_options.append('outfit')
    if '2' in option:
        selected_options.append('accessories')
    if '3' in option:
        selected_options.append('shoes')
    if '4' in option:
        selected_options.append('handbags')

    # Add a string before the user's question for LLAVA prompt
    llava_prompt = f'''For the given input, always describe the dress and whether the dress looks like men's or women's dress. 
    Importantly, never give suggestions for all genders together. Understand whether the outfit belongs to men or women or unisex and
    give suggestions for that gender alone. For example: if the dress looks feminine, suggesting a good chain or earrings is desirable. Compress your suggestions to 3 concise lines maximum and be very clear.
    Give as much visual representations as possible: {user_input}

    User selected options: {", ".join(selected_options)}'''

    if first_image_link is None:
        # Read and encode the first image in base64
        first_image_path = os.getcwd() + '\\dress_B0006HJ91S.jpg'
        with open(first_image_path, 'rb') as first_image_file:
            first_image_data = base64.b64encode(first_image_file.read()).decode('utf-8')
        first_image_link = first_image_data  # Store the link for the first image

    # Apply previous modifications to the LLAVA prompt
    modified_prompt = llava_prompt
    for mod in modifications:
        modified_prompt += f"\n{mod}"

    # Payload data for LLAVA model with image input
    llava_payload = {
        "model": "llava:7b-v1.6-mistral-q5_K_M",
        "prompt": modified_prompt,
        "stream": False,
        "images": [first_image_link]  # Use the link of the first image as input
    }

    # Headers with Hugging Face token for authentication
    headers = {
        'Authorization': f'Bearer {hugging_face_token}'
    }

    # Send POST request to LLAVA model with authentication headers
    llava_response = requests.post(llava_url, json=llava_payload, headers=headers)

    # Check response status from LLAVA model
    if llava_response.status_code == 200:
        # Get the LLAVA model's response
        llava_response_data = llava_response.json()
        llava_suggestion = llava_response_data['response']
        print(llava_response_data['response'])
        # Check if the user wants to change the dress based on LLAVA suggestion
        change_dress_input = input("Do you want to change your dress according to LLAVA's suggestion? (yes/no): ")

        if change_dress_input.lower() == 'yes':
            change_dress = True

        if change_dress and llava_suggestion:
            # Check Diffusion model availability
            if check_model_availability():
                # Payload data for Diffusion model
                diffusion_payload = {
                    "inputs": llava_suggestion
                }

                # Send POST request to Diffusion model
                diffusion_response = requests.post(runwayml_url, json=diffusion_payload, headers=headers)

                if diffusion_response.status_code == 200:
                    print("Image generated successfully!")

                    # Display the generated image
                    generated_image = Image.open(BytesIO(diffusion_response.content))
                    generated_image.show()

                else:
                    print(f"Error generating image with Diffusion model: {diffusion_response.status_code} - {diffusion_response.text}")

            else:
                print("Diffusion model is not available. Showing LLAVA suggestions only.")

    else:
        print(f"Error accessing the LLAVA model: {llava_response.status_code} - {llava_response.text}")

    # Wait for a few seconds before next iteration to avoid frequent requests
    time.sleep(5)
```

# Find Similar dress

```python
import base64
import requests
import csv
import os

# API endpoint URL for LLAVA
llava_url = 'http://8.12.5.48:11434/api/generate'

# Your Hugging Face token for authentication
hugging_face_token = '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'
base_url = os.getcwd()

# Load IDs and image links from a text file with different encodings
def load_image_data(file_path):
    encodings = ['utf-8', 'latin-1']  # Try different encodings
    data = []
    for encoding in encodings:
        try:
            with open(file_path, 'r', encoding=encoding) as file:
                for line in file:
                    parts = line.strip().split('\t')
                    if len(parts) == 2:
                        image_id, image_link = parts
                        image_link = image_link.strip()
                        data.append({'id': image_id, 'link': image_link})
            break  # Stop trying encodings if successful
        except UnicodeDecodeError:
            continue  # Try next encoding if decoding fails
    return data

# Generate LLAVA description for an image from URL or local file path
def generate_llava_description(image_source):
    try:
        if image_source.startswith('http://') or image_source.startswith('https://'):
            # Fetch the image from URL
            response = requests.get(image_source)
            response.raise_for_status()  # Raise an error if response status is not 200
            # Encode the image in base64
            image_data = base64.b64encode(response.content).decode('utf-8')
        else:
            # Read and encode the local image in base64
            with open(image_source, 'rb') as image_file:
                image_data = base64.b64encode(image_file.read()).decode('utf-8')

        # Payload data for LLAVA model with image input
        llava_payload = {
            "model": "llava:7b-v1.6-mistral-q5_K_M",
            "prompt": f"Generate a description for this image: {image_source}",
            "stream": False,
            "images": [image_data]
        }

        # Headers with Hugging Face token for authentication
        headers = {
            'Authorization': f'Bearer {hugging_face_token}'
        }

        # Send POST request to LLAVA model with authentication headers
        llava_response = requests.post(llava_url, json=llava_payload, headers=headers)
        llava_response.raise_for_status()  # Raise an error if response status is not 200

        llava_response_data = llava_response.json()
        description = llava_response_data['response']
        return description

    except requests.exceptions.RequestException as e:
        print(f"Error processing image: {e}")
        return None

# Load image data from the text file
image_data_file = base_url + '\\image_url\\images_txt.txt'  # Update with your file path
image_data = load_image_data(image_data_file)
#print("Image data: \n", image_data)

# Load existing descriptions from CSV if available
existing_descriptions = {}
csv_file = base_url + '\\image_descriptions.csv'
if os.path.isfile(csv_file):
    with open(csv_file, 'r', newline='', encoding='latin1') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            if 'id' in row and 'link' in row and 'description' in row:
                existing_descriptions[row['id']] = {'link': row['link'], 'description': row['description']}

# Generate LLAVA descriptions for images and store/update them in CSV
with open(csv_file, 'a', newline='', encoding='latin1') as csvfile:
    fieldnames = ['id', 'link', 'description']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    for item in image_data:
        image_id = item['id']
        image_link = item['link']
        if image_id not in existing_descriptions:
            print(f"Generating LLAVA description for image with ID: {image_id}")
            description = generate_llava_description(image_link)
            if description:
                existing_descriptions[image_id] = {'link': image_link, 'description': description}
                writer.writerow({'id': image_id, 'link': image_link, 'description': description})
                print(f"LLAVA description saved for image with ID: {image_id}")

# User input image (local file path or URL)
user_image_source = base_url + '\\dress_B0006HJ91S.jpg'  # Update with your image path or URL
user_image_description = generate_llava_description(user_image_source)

if user_image_description:
    print(f"LLAVA description for user input image: {user_image_description}")
else:
    print("Failed to generate LLAVA description for user input image.")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Preprocess the description text
def preprocess_text(text):
    # Remove non-alphanumeric characters
    text = ''.join(e for e in text if e.isalnum() or e.isspace())
    # Convert to lowercase and tokenize
    tokens = text.lower().split()
    return ' '.join(tokens)

# Function to calculate cosine similarity between two vectors
def calculate_similarity(description1, description2):
    # Use TF-IDF vectorization for numerical representation
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([description1, description2])
    # Calculate cosine similarity
    similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
    return similarity

# User image description (replace with actual description)
#user_description = "Your user image description here"

most_similar_image = None
highest_similarity = 0.0

# Iterate through existing descriptions and find the most similar image
for image_id, data in existing_descriptions.items():
    existing_description = preprocess_text(data['description'])
    user_description_preprocessed = preprocess_text(user_image_description)
    similarity = calculate_similarity(existing_description, user_description_preprocessed)
    if similarity > highest_similarity:
        highest_similarity = similarity
        most_similar_image = {'id': image_id, 'link': data['link'], 'description': existing_description}

# Print the most similar image information
if most_similar_image:
    print(f"The most similar image is ID: {most_similar_image['id']}, Link: {most_similar_image['link']}")
    print(f"Similarity score: {highest_similarity}")
else:
    print("No similar image found.")
```
